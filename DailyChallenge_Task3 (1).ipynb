```python
import streamlit as st
import time

def main():
    st.set_page_config(
        page_title="Smart Arithmetic Calculator",
        page_icon="ğŸ§®",
        layout="centered"
    )

    # Header with styling
    st.title("ğŸ§® Smart Arithmetic Calculator ğŸ§®")
    st.markdown("---")

    col1, col2 = st.columns(2)

    with col1:
        operation = st.selectbox(
            "Choose Operation:",
            [
                "Addition (+)",
                "Subtraction (-)",
                "Multiplication (*)",
                "Division (/)"
            ]
        )

    with col2:
        num1 = st.number_input("Enter first number:", step=0.1)
        num2 = st.number_input("Enter second number:", step=0.1)

    # Calculate button
    if st.button("Calculate", type="primary"):
        with st.spinner("ğŸ§® Calculating..."):
            time.sleep(0.5)

            try:
                if "Addition" in operation:
                    result = num1 + num2
                    operator = "+"
                elif "Subtraction" in operation:
                    result = num1 - num2
                    operator = "-"
                elif "Multiplication" in operation:
                    result = num1 * num2
                    operator = "*"
                elif "Division" in operation:
                    if num2 == 0:
                        st.error("âŒ Cannot divide by zero!")
                        return
                    result = num1 / num2
                    operator = "/"

                # Round to 2 decimal places
                result = round(result, 2)

                # Display result in a success box
                st.success(f"{num1} {operator} {num2} = {result}")

                # Additional info in expandable section
                with st.expander("See calculation details"):
                    st.write(f"Operation: {operation}")
                    st.write(f"First number: {num1}")
                    st.write(f"Second number: {num2}")
                    st.write(f"Result (rounded to 2 decimal places): {result}")

            except Exception as e:
                st.error(f"âŒ Error: {e}")

    # Add some information at the bottom
    st.markdown("---")
    st.markdown(" Created with â¤ï¸ by Tahira Ibrahim")

if __name__ == "__main__":
    main()
```
